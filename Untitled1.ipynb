```python
df = pd.read_csv("data/경기도소방관서1인당담당인구현황.csv", encoding='CP949')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 사용 가능한 폰트 확인
# [f.name for f in fm.fontManager.ttflist if 'Nanum' in f.name]

# 한국어 출력을 위한 폰트 설정
plt.rc('font', family='NanumGothic') 

# 마이너스 기호가 깨지는 것을 방지
plt.rcParams['axes.unicode_minus'] = False

# 관서별 1인당 담당 인구 내림차순으로 정렬
df_grouped_by_station = df.groupby('관서명')['1인당담당인구(명)'].mean().reset_index()
df_sorted = df_grouped_by_station.sort_values('1인당담당인구(명)', ascending=False)

# 그래프 가로로 넓게 설정
plt.figure(figsize=(10, 8))

# 관서별 1인당 담당 인구 바 그래프 생성, x축과 y축 변경
sns.barplot(x='관서명', y='1인당담당인구(명)', data=df_sorted)

# x축 레이블 회전
plt.xticks(rotation=45)

plt.ylabel('1인당 담당 인구(명)')
plt.xlabel('관서명')
plt.title('관서별 1인당 담당 인구, 내림차순 정렬')
# plt.savefig('관서별 1인당 담당 인구.png')
plt.show()
```
