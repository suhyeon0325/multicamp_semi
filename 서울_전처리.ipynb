```python
import geopandas as gpd

# GeoDataFrame 생성
gdf_grid = gpd.GeoDataFrame(seoul_grid, geometry='geometry')
gdf_merged = gpd.GeoDataFrame(seoul_merged, geometry='geometry')

# sjoin을 사용하여 right join 수행
gdf_joined = gpd.sjoin(gdf_merged, gdf_grid, how='inner', op='intersects')
```

```python
# 'geometry' 열의 고유값 개수 확인
unique_geometries = gdf_cleaned['geometry'].nunique()
print(f"'geometry' 열의 고유값 개수: {unique_geometries}")

# 전체 데이터 개수에서 'geometry' 열의 고유값 개수를 빼서 중복된 'geometry' 값의 개수를 확인
duplicate_geometries = len(gdf_cleaned) - unique_geometries
print(f"'geometry' 열에서 중복된 값의 개수: {duplicate_geometries}")
```

```python
import pandas as pd
import geopandas as gpd

def select_representative_row(group):
    """
    각 geometry 그룹에 대해 비어 있지 않은 'pop', 'total_b', 'res_Single' 값이 
    가장 많은 행을 선택합니다. 동일한 수의 비어 있지 않은 값이 있는 경우,
    첫 번째 행을 선택합니다.
    """
    # 'pop', 'total_b', 'res_Single' 열에서 비어 있지 않은 값의 개수 계산
    info_columns = ['pop', 'total_b', 'res_single', 'res_multi', 'b_over_20']
    group['non_empty_count'] = group[info_columns].notnull().sum(axis=1)
    
    # 가장 많은 비어 있지 않은 값을 가진 행 선택
    representative_row = group.sort_values('non_empty_count', ascending=False).iloc[0]
    
    return representative_row

# 'geometry' 열을 기준으로 그룹화하고, 각 그룹에 대해 대표 행을 선택
representative_rows = gdf_joined.groupby('geometry').apply(select_representative_row)

# 대표 행들만 포함하는 새 GeoDataFrame 생성
gdf_representative = gpd.GeoDataFrame(representative_rows, geometry='geometry').reset_index(drop=True)

# 결과 확인

```
