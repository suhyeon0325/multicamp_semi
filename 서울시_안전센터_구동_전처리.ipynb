```python
gdf = gpd.read_file("data/동경계/동경계_geo.shp")
```

```python
df = pd.read_csv("data/서울시_소방서_안전세터_구조대.csv")
```

```python
df
```

```python
geometry = [Point(xy) for xy in zip(df['경도'], df['위도'])]
df_geo = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

joined_df = gpd.sjoin(df_geo, gdf, how="left", op='within')

result_df = joined_df[['서ㆍ센터명', '유형구분명', '위도', '경도', '구', '동']]
```
