```python
df_filtered.drop(['화재유형명', '현장소방지역대거리', '화재발생연도'], axis=1, inplace=True)
```

```python
# 계절&낮밤 구분
def get_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    elif month in [12, 1, 2]:
        return '겨울'

def get_day_period(hour):
    return '낮' if 6 <= hour <= 18 else '밤'

df_filtered['계절'] = df_filtered['화재발생월'].apply(get_season)
df_filtered['시간대'] = df_filtered['화재발생시'].apply(get_day_period)
```

```python
# 날짜, 시간 열 합치기
df_filtered['화재발생일시'] = pd.to_datetime('2021/' +
                                           df_filtered['화재발생월'].astype(str) + '/' +
                                           df_filtered['화재발생일'].astype(str) + ' ' +
                                           df_filtered['화재발생시'].astype(str) + ':' +
                                           df_filtered['화재발생분'].astype(str),
                                           format='%Y/%m/%d %H:%M')
```

```python
# 평일 주말 구분
df_filtered['주말여부'] = df_filtered['요일'].apply(lambda x: '주말' if x in ['토', '일'] else '평일')
```

```python
# 팝업 텍스트를 생성하는 함수 (HTML 스타일 적용)
def create_popup_html(row):
    return f'''
    <html>
        <head><style>
            .popup {{
                font-family: Arial, sans-serif;
                font-size: 12px;
                color: #333333;
            }}
            .title {{
                font-weight: bold;
                color: #0078A8;
                margin-bottom: 5px;
            }}
            .info {{
                margin-bottom: 2px;
            }}
        </style></head>
        <body>
            <div class="popup">
                <div class="title">화재 정보</div>
                <div class="info">사망수: {row['사망수']}, 부상자수: {row['부상자수']}</div>
                <div class="info">재산피해금액: {row['재산피해금액']}원</div>
                <div class="info">출동소요시간: {row['출동소요시간']}초</div>
                <div class="info">화재진압시간: {row['화재진압시간']}초</div>
                <div class="info">위치: {row['시군구명']}, {row['읍면동명']}</div>
                <div class="info">계절: {row['계절']}, 시간대: {row['시간대']}</div>
                <div class="info">화재발생일시: {row['화재발생일시']}</div>
            </div>
        </body>
    </html>
    '''
```

```python
import folium

# NaN 값을 가진 행 제거
df = df_filtered.dropna(subset=['위도', '경도'])

# 계절별 색상을 결정하는 함수
def get_color(season):
    if season == '봄':
        return 'green'
    elif season == '여름':
        return 'red'
    elif season == '가을':
        return 'orange'
    elif season == '겨울':
        return 'blue'
    else:
        return 'gray'  # 계절 정보가 없는 경우


# 서울의 중심 좌표로 지도 생성
map_seoul = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# 데이터프레임을 순회하며 CircleMarker 추가
for idx, row in df.iterrows():
    color = get_color(row['계절'])  # 계절별 색상
    tooltip_text = f'출동소요시간: {row["출동소요시간"]}초'  # 툴팁 텍스트
    popup_html = create_popup_html(row)  # 팝업 HTML 텍스트 생성
    popup = folium.Popup(popup_html, max_width=300)  # 여기서 Popup 객체 생성

    folium.CircleMarker(
        [row['위도'], row['경도']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        tooltip=tooltip_text,
        popup=popup  # 생성된 Popup 객체를 사용
    ).add_to(map_seoul)

# 지도 저장 또는 표시
map_seoul
```
