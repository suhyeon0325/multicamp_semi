```python
import pandas as pd
```

```python
df_2022 = pd.read_csv("data/화재발생_동별_2022.csv", encoding='cp949')
df_2021 = pd.read_csv("data/화재발생_동별_2021.csv", encoding='cp949')
```

```python
df_2022.head()
```

```python
df_2021.head()
```

```python
# '-'를 0으로 변환
df_2021 = df_2021.replace('-', 0)
df_2022 = df_2022.replace('-', 0)
```

```python
# 데이터 타입을 int로 변환
for col in df_2021.columns[2:]:
    df_2021[col] = df_2021[col].astype(int)
    df_2022[col] = df_2022[col].astype(int)
```

```python
# 데이터 합치기
df_combined = pd.concat([df_2021, df_2022])
```

```python
# '자치구'와 '동' 기준으로 그룹화하여 합계 계산
df_sum = df_combined.groupby(['자치구', '동']).sum().reset_index()
```

```python
df_sum
```

```python
# 각 '구'별로 합계를 계산
df_group_sum = df_sum.groupby('자치구').sum().reset_index()

# '동' 열에 '전체'를 추가
df_group_sum['동'] = df_group_sum['자치구'] + ' 전체'

# 원래의 데이터프레임에 합계 데이터 추가
df_sum_with_total = pd.concat([df_sum, df_group_sum], ignore_index=True)

# 결과 확인
df_sum_with_total
```

```python
df_sum_with_total.to_csv("data/동별_화재발생_장소_2021_2022.csv")
```
